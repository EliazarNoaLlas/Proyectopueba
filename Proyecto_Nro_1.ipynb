## Ejercicio N¬∫ 2

Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso m√≠nimo, m√°ximo y media. 
El resultado final debe ser una serie con los estad√≠sticos.


```python
import pandas as pd

# Crear un diccionario con los ingresos mensuales de los clientes de BBVA
ingresos = {'Enero': [5000, 6000, 7000],
            'Febrero': [8000, 9000, 10000],
            'Marzo': [11000, 12000, 13000],
            'Abril': [14000, 15000, 16000],
            'Mayo': [17000, 18000, 19000],
            'Junio': [20000, 21000, 22000],
            'Julio': [23000, 24000, 25000],
            'Agosto': [26000, 27000, 28000],
            'Septiembre': [29000, 30000, 31000],
            'Octubre': [32000, 33000, 34000],
            'Noviembre': [35000, 36000, 37000],
            'Diciembre': [38000, 39000, 40000]}

# Crear un DataFrame con los ingresos de todos los clientes, indexado por mes
df_ingresos = pd.DataFrame(ingresos, index=['Juan Perez', 'Maria Torres', 'Pedro Diaz'])

# Calcular el ingreso m√≠nimo, m√°ximo y media de cada mes
ingreso_minimo = df_ingresos.min(axis=0)
ingreso_maximo = df_ingresos.max(axis=0)
ingreso_media = df_ingresos.mean(axis=0)

# Crear un nuevo DataFrame con los estad√≠sticos calculados, indexado por mes
df_estadisticos = pd.DataFrame({'M√≠nimo': ingreso_minimo, 'M√°ximo': ingreso_maximo, 'Media': ingreso_media})

# Mostrar el DataFrame resultante por pantalla
print(df_estadisticos)
```

## Ejercicio N¬∫ 3

 Escribir programa que genere y muestre por pantalla un DataFrame con los datos de la siguiente tabla: 

Mes Ventas Gastos 
Enero 30500 22000 
Febrero 35600 23450 
Marzo 28300 18100 
Abril 33900 35700 
Mayo 42500 32450 
A continuaci√≥n, generar nuevas columna: 
1. Columna 1 = Gasto/Ventas*100. 
2. Columna 2 = Si Ventas>= 30000, imprimir ‚ÄúMeta Superada‚Äù, caso contrario imprimir ‚ÄúMeta no Superada‚Äù. 
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto. 4. Obtener el total de las ventas de los √∫ltimos 3 meses 5. Obtener el total de gastos de febrero a abril



### FORMA 1

```python
import pandas as pd

# Creamos el DataFrame con los datos de la tabla
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
        'Ventas': [30500, 35600, 28300, 33900, 42500],
        'Gastos': [22000, 23450, 18100, 35700, 32450]}
df = pd.DataFrame(data)

# A√±adimos la columna del gasto/ventas*100
df['% Gasto/Ventas'] = df['Gastos'] / df['Ventas'] * 100

# A√±adimos la columna para verificar si la meta fue superada
df['Meta'] = df['Ventas'].apply(lambda x: 'Meta Superada' if x >= 30000 else 'Meta no Superada')

# A√±adimos la columna del sobregasto si es que lo hay
df['Sobregasto'] = df.apply(lambda x: x['Gastos'] - x['Ventas'] if x['Gastos'] > x['Ventas'] else 0, axis=1)

# Obtenemos el total de las ventas de los √∫ltimos 3 meses
total_ventas_ultimos_3_meses = df.tail(3)['Ventas'].sum()
print(f"Total ventas √∫ltimos 3 meses: {total_ventas_ultimos_3_meses}")

# Obtenemos el total de gastos de febrero a abril
total_gastos_feb_abr = df.loc[(df['Mes'] == 'Febrero') | (df['Mes'] == 'Marzo') | (df['Mes'] == 'Abril'), 'Gastos'].sum()
print(f"Total gastos febrero a abril: {total_gastos_feb_abr}")

# Mostramos el DataFrame final
df
```

### FORMA 2

```python
import pandas as pd
import numpy as np

# Creamos el DataFrame con los datos de la tabla
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
        'Ventas': [30500, 35600, 28300, 33900, 42500],
        'Gastos': [22000, 23450, 18100, 35700, 32450]}
df = pd.DataFrame(data)

# A√±adimos la columna del gasto/ventas*100
df['% Gasto/Ventas'] = np.round(df['Gastos'] / df['Ventas'] * 100, 2)

# A√±adimos la columna para verificar si la meta fue superada
df['Meta'] = np.where(df['Ventas'] >= 30000, 'Meta Superada', 'Meta no Superada')

# A√±adimos la columna del sobregasto si es que lo hay
df['Sobregasto'] = np.where(df['Gastos'] > df['Ventas'], df['Gastos'] - df['Ventas'], 'No hay sobregastos')

# Obtenemos el total de las ventas de los √∫ltimos 3 meses
total_ventas_ultimos_3_meses = df['Ventas'].tail(3).sum()
print(f"Total ventas √∫ltimos 3 meses: {total_ventas_ultimos_3_meses}")

# Obtenemos el total de gastos de febrero a abril
total_gastos_feb_abr = df.loc[df['Mes'].isin(['Febrero', 'Marzo', 'Abril']), 'Gastos'].sum()
print(f"Total gastos febrero a abril: {total_gastos_feb_abr}")

# Mostramos el DataFrame final
df
```
